```julia
#--------------------------------------------
# Run functions and save results by scenarios
#--------------------------------------------

# Packages
using JuMP
using Gurobi
using DataFrames
using XLSX
using CSV
```

```julia
include("flex_scheduling.jl")
include("strc_scheduling.jl")
```

```julia
# Scenario A, July 1, 2024
date = 1
N_stack_A = 1 # number of electrolyzer stacks
C_stack_A = 10 # capacity of individual electrolyzer stack in MW
Ramping = [0.2, 0.5, 1.0]

ramping_baseline = 0
baseline_false = false
baseline_true = true
stack_plot = false

data_A = DataFrame()
cost_A = Dict()

# Baseline operation
T, lambda_M, total_p_prof, s_prof, p_prof, optimal_cost = flex_scheduling(date, N_stack_A, C_stack_A, ramping_baseline, baseline_true, stack_plot)
data_A[!, "time step"] = T
data_A[!, :RTM] = lambda_M
data_A[!, "p: baseline"] = total_p_prof
data_A[!, "s: baseline"] = s_prof
cost_A["baseline"] = optimal_cost

# Flexible operation with different ramping
for r in Ramping
    T, lambda_M, total_p_prof, s_prof, p_prof, optimal_cost = flex_scheduling(date, N_stack_A, C_stack_A, r, baseline_false, stack_plot)
    data_A[!,"p: A (r = $r)"] = total_p_prof
    data_A[!,"s: A (r = $r)"] = s_prof
    cost_A["A (r = $r)"] = optimal_cost
end
```

```julia
# Scenario B, July 1, 2024
date = 1
N_stack_B = 10 # number of electrolyzer stacks
C_stack_B = 1 # capacity of individual electrolyzer stack in MW
Ramping = [0.2, 0.5, 1.0]

ramping_baseline = 0
baseline_false = false
baseline_true = true
stack_plot = false

data_B = DataFrame()
cost_B = Dict()

# Baseline operation
T, lambda_M, total_p_prof, s_prof, p_prof, optimal_cost = flex_scheduling(date, N_stack_B, C_stack_B, ramping_baseline, baseline_true, stack_plot)
data_B[!, "time step"] = T
data_B[!, :RTM] = lambda_M
data_B[!, "p: baseline"] = total_p_prof
data_B[!, "s: baseline"] = s_prof
cost_B["baseline"] = optimal_cost

# Flexible operation with different ramping
for r in Ramping
    T, lambda_M, total_p_prof, s_prof, p_prof, optimal_cost = flex_scheduling(date, N_stack_B, C_stack_B, r, baseline_false, stack_plot)
    data_B[!, "p: B (r = $r, total)"] = total_p_prof
    data_B[!, "s: B (r = $r)"] = s_prof
    for k in 1:N_stack_B
        data_B[!, "p: B (r = $r, k = $k)"] = p_prof[k]
    end
    cost_B["B (r = $r)"] = optimal_cost
end
```

```julia
# Scenario C, July 1, 2024
date = 1
N_stack_C = 10 # number of electrolyzer stacks
C_stack_C = 1 # capacity of individual electrolyzer stack in MW
Protocol = ["v1", "v2", "v3"] # details of each set of protocols are described in data.jl
Minmax_mode = [true, false]
Oper_mode = ["minmax", "protocol"]

ramping_baseline = 0
baseline_true = true
minmax_false = false
stack_plot = false

data_C = DataFrame()
cost_C = Dict()

# Baseline operation
T, lambda_M, total_p_prof, s_prof, p_prof, optimal_cost = flex_scheduling(date, N_stack_C, C_stack_C, ramping_baseline, baseline_true, stack_plot)
data_C[!, "time step"] = T
data_C[!, :RTM] = lambda_M
data_C[!, "p: baseline"] = total_p_prof
data_C[!, "s: baseline"] = s_prof
cost_C["baseline"] = optimal_cost

for (idx_v, v) in enumerate(Protocol)
    for (idx_m, m) in enumerate(Minmax_mode)
        T, lambda_M, total_p_prof, s_prof, p_prof, optimal_cost = strc_scheduling(date, N_stack_C, C_stack_C, v, m, stack_plot)
        data_C[!, "p: C (v = $idx_v, $(Oper_mode[idx_m]))"] = total_p_prof
        data_C[!, "s: C (v = $idx_v, $(Oper_mode[idx_m]))"] = s_prof
        for k in 1:N_stack_C
            data_C[!, "p: C (v = $idx_v, $(Oper_mode[idx_m]),k = $k)"] = p_prof[k]
        end    
        cost_C["C (v = $idx_v, m = $(Oper_mode[idx_m]))"] = optimal_cost
    end
end
```

```julia
# Report costs for all scenarios

cost_all = sort(merge(cost_A, cost_B, cost_C)) # sorted by keys
cost_all_df = DataFrame(cost_all)
#XLSX.writetable("results.xlsx", "Scen A" => data_A, "Scen B" => data_B, "Scen C" => data_C, "Opt Cost" => cost_all_df; overwrite= true)

cost_all
```
