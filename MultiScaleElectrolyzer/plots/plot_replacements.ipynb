```julia

using Plots
using LaTeXStrings
using Dates


plot_font = "Computer Modern"
default(fontfamily=plot_font,
        linewidth=2, framestyle=:box, label=nothing, grid=false)
# scalefontsizes(1.2)

function plot_all_scenarios(hub_avg_data, hub_avg_data_const, pan_data, pan_data_const, dist_data_const)

    hours = 1:length(hub_avg_data["A"])
    # values = rand(length(hours))
    # Create the Time Axis
    # We pick a start date (e.g., Jan 1, 2024).
    # Adding `Hour.(hours)` creates a vector of DateTime objects.
    time_axis = DateTime(2014, 1, 1) .+ Hour.(hours)
    xaxislabels = (DateTime(2014, 1, 1):Year.(1):DateTime(2036, 1, 1), Dates.format.(DateTime(2014, 1, 1):Year.(1):DateTime(2036, 1, 1), "yyyy"))

    p = plot(time_axis, hub_avg_data["A"].*33.36./10;
        label = "Hub Average, Flexible Operation",
        xlabel="Time (Year)", ylabel="Efficiency (% LHV)", legend=:outertop,
        xticks= xaxislabels, xrotation=45, ylims= (54, 66), yticks=55:1:65)


    plot!(time_axis, hub_avg_data_const["A"].*3.336, label = "Hub Average, Non-Flexible Operation")

    plot!(time_axis, pan_data["A"].*3.336, label = "Panhandle, Flexible Operation")

    plot!(time_axis, pan_data_const["A"].*3.336, label = "Panhandle, Non-Flexible Operation")

    plot!(time_axis, dist_data_const["A"].*3.336, label = "Distribution, Non-Flexible Operation")

    # plot!(time_axis, dist_data_const["A"].*3.336, label = "Constant Operation")
    # plot!(time_axis, heuristic["A"].*33.36./10;
        # label = "Heuristic Degradation and Optimal Replacement")
    # plot!(time_axis, nodeg["A"].*33.36./10, label = "No Degradation")
    # plot!(time_axis, fullheuristic["A"].*33.36/10, label = "Heuristic Degradation and Replacement")
    plot!(time_axis, ones(length(hub_avg_data["A"]))*55, linestyle=:dash,color=:gray, label=false)
    annotate!(p, DateTime(2018,3,1), 55.5, text("Replacement Heuristic (10% Efficiency Loss)",plot_font,7, :gray))
end

function plot_flex_nonflex(hub_avg_data, hub_avg_data_const)
    
    hours = 1:length(hub_avg_data["A"])
    # values = rand(length(hours))
    # Create the Time Axis
    # We pick a start date (e.g., Jan 1, 2024).
    # Adding `Hour.(hours)` creates a vector of DateTime objects.
    time_axis = DateTime(2014, 1, 1) .+ Hour.(hours)
    xaxislabels = (DateTime(2014, 1, 1):Year.(1):DateTime(2036, 1, 1), Dates.format.(DateTime(2014, 1, 1):Year.(1):DateTime(2036, 1, 1), "yyyy"))

    p = plot(time_axis, hub_avg_data["A"].*33.36./10;
        label = "Flexible Operation",
        xlabel="Time (Year)", ylabel="Efficiency (% LHV)", legend=:outertop, color=:crimson,
        xticks= xaxislabels, xrotation=45, ylims= (54, 66), yticks=55:1:65)


    plot!(time_axis, hub_avg_data_const["A"].*3.336, label = " Non-Flexible Operation", color=:steelblue)


    # plot!(time_axis, dist_data_const["A"].*3.336, label = "Constant Operation")
    # plot!(time_axis, heuristic["A"].*33.36./10;
        # label = "Heuristic Degradation and Optimal Replacement")
    # plot!(time_axis, nodeg["A"].*33.36./10, label = "No Degradation")
    # plot!(time_axis, fullheuristic["A"].*33.36/10, label = "Heuristic Degradation and Replacement")
    plot!(time_axis, ones(length(hub_avg_data["A"]))*55, linestyle=:dash,color=:gray, label=false)
    annotate!(p, DateTime(2018,3,1), 55.5, text("Replacement Heuristic (10% Efficiency Loss)",plot_font,7, :gray))

    plot_path = (@__DIR__)*"/../plots/"
    savefig(p, plot_path*"replacements.pdf")
end


plot_flex_nonflex(hub_avg_data, hub_avg_data_const)

```
